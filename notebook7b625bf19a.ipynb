```python
# print firsst 10 fibonacci numbers
def fibonacci(n):
    a,b=0,1
    for i in range(n):
        print(a,end=" ")
        a,b=b,a+b
fibonacci(10)
```

```python
# check whether a number is prime
num=int(input("enter number:"))
if num<=1:
    print("not a prime number")
else:
  is_prime=True
  for i in range(2,num):
      if num % i==0:
          is_prime=False
          break
  if is_prime:
      print("prime number")
  else:
      print("not a prime number")
```
